```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By         
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
# -*- coding: utf-8 -*-
"""
Created on Thu Oct 12 16:23:29 2023 
"""
from bs4 import BeautifulSoup
import pandas as pd
import os

```

```python
chrome_options = Options()
driver = webdriver.Chrome(service=Service(), options=chrome_options)
```

```python
filter = "?page_number=1&page_size=100"
driver.get("https://www.kabum.com.br/busca/cpu" + filter)

soup = BeautifulSoup(driver.page_source, 'html.parser')
elements = soup.find_all('div')
# Obtener todas las clases usadas en la página
todas_clases = set()

for tag in soup.find_all(True):  # True = todos los tags
    clases = tag.get('class')
    if clases:
        todas_clases.update(clases)
```

```python
todas_clases
```

```python
cards = soup.find_all('div', class_='rounded-4')
productos = []

for card in cards:
    try:
        name = card.find('h3').text.strip()
        price = card.find('span', class_='priceCard').text.strip()
        image_url = card.find('img')['src']
        product_url = card.find('a')['href']
        productos.append({
            'name': name,
            'price': price,
            'image_url': image_url,
            'product_url': product_url
        })
    except AttributeError:
        continue  # Si falta algún dato, pasa al siguiente

productos
```
